```python
import os
import pickle
from pprint import pprint

# parameters you already have
func_name   = "StybTang"
kernel_name = "gcauchy"
kernel_base = "gcauchy"
seed        = 1

# build the relative path
fn = os.path.join(
    "hyperparams",
    func_name,
    kernel_name,
    f"hyperparams_{func_name}_{kernel_base}_seed{seed}.pkl"
)

# load the pickle
with open(fn, "rb") as f:
    loaded = pickle.load(f)
```

```python
# 1. check type
print(type(loaded))

# 2. if it’s a dict (most common), look at its keys
if isinstance(loaded, dict):
    print("keys:", loaded.keys())
    # and pretty-print the contents
    pprint(loaded)

# 3. for arbitrary objects you can dir() them
else:
    print(dir(loaded))
    pprint(vars(loaded) if hasattr(loaded, "__dict__") else loaded)
```
